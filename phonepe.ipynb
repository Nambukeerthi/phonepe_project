```python
import os
import json
import psycopg2
import numpy as np
import pandas as pd
```

```python
#1 aggregatte transaction
# get sourse file details 
path1 = "C:/Users/Raj/Desktop/dsproject/phonepedata/aggregated/transaction/country/india/state/" #changed to right slas
aggre_list = os.listdir(path1)


columns1 = {"States": [], "Years":[], "Quarter": [], "Transaction_type": [],"Transaction_count" :[],  "Transaction_amount":[]}

for state in aggre_list:
    cur_state = path1 + state + "/"   # right slas for open folders
    #print(curstate) - print all state folders
    aggre_year_list = os.listdir(cur_state)
    #print(aggre_year_list) - print all year folders in state folder
    
    for year in aggre_year_list:
        cur_year = cur_state + year +"/"  # right slas for open folders
        #print(cur_year) - print 
        aggre_file_list = os.listdir(cur_year)
        #print(aggre_file_list) - print all jason files
        
        for file in aggre_file_list:
            cur_file = cur_year + file 
            data = open(cur_file,"r")
            
            A = json.load(data)
            
            for i in A["data"]["transactionData"]:
                name = i["name"]
                count = i["paymentInstruments"][0]["count"] 
                amount = i["paymentInstruments"][0]["amount"]
                columns1["Transaction_type"].append(name)
                columns1["Transaction_count"].append(count)
                columns1["Transaction_amount"].append(amount)
                columns1["States"].append(state)
                columns1["Years"].append(year)
                columns1["Quarter"].append(int(file.strip(".json")))
```
